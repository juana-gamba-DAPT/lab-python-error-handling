def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
#available products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# inventary
def initialize_inventory(products):
    inventory = {}
    for product in products:
        # Pedir al usuario la cantidad de cada producto
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    print("Please enter a positive number.")
                else:
                    inventory[product] = quantity
                    break
            except ValueError:
                print("Please enter a valid number.")
    return inventory

inventory = initialize_inventory(products)

# To get Orders
def get_customer_orders():
    customer_orders = []
    while True:
        try:
            n = int(input("How many orders do you want to take? "))
            if n < 1:
                print("Please enter at least one order.")
            else:
                break
        except ValueError:
            print("Please enter a valid number.")
    
    for order in range(n):
        while True:
            product = input("Enter a product to order: ").lower()
            if product in inventory:
                customer_orders.append(product)
                break
            else:
                print(f"Sorry, {product} is not available. Please choose from the following: {', '.join(inventory.keys())}")
    return customer_orders

customer_orders = get_customer_orders()

#total preu
def calculate_total_price(customer_orders):
    unit_price = {}
    total_price = 0
    for product in customer_orders:
        while True:
            try:
                price = int(input(f"Please insert the unit price of the {product}: "))
                if price < 0:
                    print("Please enter a positive price.")
                else:
                    unit_price[product] = price
                    total_price += price
                    break
            except ValueError:
                print("Please enter a valid price.")
    
    print(f"Unit prices for the products: {unit_price}")
    return total_price

total = calculate_total_price(customer_orders)
print(f"Total price: {total}")

#new inventary
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if inventory[product] > 0:
            inventory[product] -= 1
        else:
            print(f"Sorry, {product} is out of stock.")
    
    # Eliminar productos con cantidad 0 en el inventario
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    
    print("Updated inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
    
    return inventory

# Actualizar inventario
inventory = update_inventory(customer_orders, inventory)
````
